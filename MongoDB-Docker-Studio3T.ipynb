### Question 1

#### Step 1
For simplifying the query related to question 5, we will create one review in the listing document, specifically the longest review among them.

- __TRANSFORMATION APPLIED:__ Creating new field ("BIGGEST_COMMENT") to store the biggest comment in a listing.

- __REASONING BEHIND:__ In question 5, the task involved frequent queries to identify the author of the longest review in New York. By pre-calculating and storing the longest comment for each listing, this simplifies the query significantly, eliminating the need for joins. This implementation optimizes computation and reduces server load. However, it's important to note that this approach can lead to stale data if not updated promptly by setting up a trigger to automatically update this field whenever a new review is added.

- __EXPECTED RESULT:__ In this scenario, having the largest comment for each listing already stored results in a simpler and faster query, specifically the query described in question 5.

```python
# Define the aggregation pipeline to find the review with the longest comment and set the BIGGEST_COMMENT field

pipeline = [
    {
        "$set": {
            # Set the BIGGEST_COMMENT field using $arrayElemAt to get the first element of the filtered reviews array
            "BIGGEST_COMMENT": {
                "$arrayElemAt": [
                    {
                        "$filter": {
                            "input": "$reviews",  # Input is the reviews array
                            "cond": {
                                "$eq": [
                                    # Compare the length of comments in each review to the maximum length
                                    { "$strLenCP": { "$ifNull": ["$$this.comments", ""] } },
                                    {
                                        "$max": {
                                            "$map": {
                                                "input": "$reviews",  # Map over the reviews array
                                                "in": { "$strLenCP": { "$ifNull": ["$$this.comments", ""] } }  # Calculate the length of comments
                                            }
                                        }
                                    }
                                ]
                            }
                        }
                    },
                    0  # Get the first (and only) element from the filtered reviews array
                ]
            }
        }
    },
    {
        "$out": "listingsAndReviews_new"  # Overwrite the collection with the transformed documents
    }
]

# Execute the aggregation pipeline to transform and update the collection
db.listingsAndReviews_new.aggregate(pipeline)
```

#### Step 3
Now we will create a new field in the listings documents with only the top 5 most recent reviews

- __TRANSFORMATION APPLIED:__ Creating new field ("TOP_REVIEWS"). Implementing the Subset Pattern.

- __REASONING BEHIND:__ We are implementing a new field dedicated to storing the five most recent reviews for specific listings. This field will be updated periodically through a query to ensure that it always reflects the latest reviews. This is one of the strategies to combat staleness. This approach optimizes the efficiency of retrieving listing documents by providing only the most current comments when queried by users.

- __EXPECTED RESULT:__  Some duplication of reviews but it optimizes performance and reduced load. Increased information efficiency within the documents.

```python
db.Reviews.find_one()
```

```python
pipeline = [
    # Sort reviews by listing_id and date in descending order
    {
        '$sort': {'listing_id': 1, 'date': -1}
    },
    # Group reviews by listing_id and push each review into an array
    {
        '$group': {
            '_id': '$listing_id',
            'reviews': {'$push': '$$ROOT'}
        }
    },
    # Project only the top 5 reviews for each listing
    {
        '$project': {
            'listing_id': '$_id',
            'TOP_REVIEWS': {'$slice': ['$reviews', 5]}
        }
    }
]

# Execute aggregation pipeline
result = list(db.Reviews.aggregate(pipeline))
pprint(result[0])
```

```python
# Update listing documents with top_reviews field
for listing_data in result:
    listing_id = listing_data['listing_id']
    top_reviews = listing_data['TOP_REVIEWS']
    
    # Update the listing document with top_reviews field
    db.listingsAndReviews_new.update_one(
        {'_id': listing_id},
        {'$set': {'TOP_REVIEWS': top_reviews}}
    )
```

Furthermore, the transaction prices are currently stored as strings. We require them to be represented as numerical values (floats) with precision limited to two decimal places.

- __Transformation Applied:__ Rounding Prices in Transaction Documents.

- __Reasoning Behind:__ Rounding the price values ensures that transaction data is uniformly formatted with a fixed number of decimal places, which enhances data quality and consistency.
  
- __Expected Result:__ The updated transaction data with rounded prices will be clearer and more usable for querying, analysis, and presentation purposes.

Before any changes

```python
document = db.Transactions.find_one()

pprint(document["transactions"][0]["price"])
pprint(type(document["transactions"][0]["price"]))
```

```python
# Variables to keep track of updated documents
updated_count = 0
total_count = 0

# Iterate over all documents in the collection
for document in db.Transactions.find():
    # Extract the transactions array from the document
    transactions = document.get('transactions', [])

    # Process each transaction and update the price
    updated_transactions = []
    for transaction in transactions:
        # Convert the price from string to float with 2 decimal places
        old_price_str = transaction.get('price', '0.0')
        try:
            old_price_float = float(old_price_str)
            new_price_float = round(old_price_float, 2)
        except ValueError:
            new_price_float = 0.0  # Default value or specific error handling

        # Create a new transaction with the updated price
        updated_transaction = {
            'date': transaction['date'],  # Keep the existing date
            'price': new_price_float  # Update the price
        }
        updated_transactions.append(updated_transaction)

    #Update the document with the modified transactions array
    result = db.Transactions.update_one(
        {'_id': document['_id']},
        {'$set': {'transactions': updated_transactions}}
    )

    if result.modified_count > 0:
        updated_count += 1

    total_count += 1

# Calculate the ratio of updated documents
if total_count > 0:
    ratio_updated = f"{updated_count}/{total_count}"
else:
    ratio_updated = "0/0"

# Print the ratio of updated documents
print(f"Documents updated successfully: {ratio_updated}")
```

#### Step 3
Check for transactions that occurred before the bucket start date or after the bucket end date. Then reset the bucket start date and bucket end date to the most recent date in the array and the latest date in that array.

Check for transactions done before the bucket start date

```python
documents = db.Transactions.find()

total = 0
before_bucket = False

for document in documents:
    bucket_start = document["bucket_start_date"]
    transactions_list = document["transactions"]
    for transaction in transactions_list:
        if transaction["date"] < bucket_start:
            before_bucket = True
    if before_bucket:
        total += 1
    before_bucket = False
print(total)
```

Check for transactions done after the bucket end date

```python
total = 0
after_bucket = False

for document in documents:
    bucket_end = document["bucket_end_date"]
    transactions_list = document["transactions"]
    for transaction in transactions_list:
        if transaction["date"] > bucket_end:
            after_bucket = True
    if after_bucket:
        total += 1
    after_bucket = False
print(total)
```

- __TRANSFORMATION APPLIED:__ Updating field values in the fields bucket_end_date and bucket_start_date.

- __REASONING BEHIND:__ After visually inspecting some documents, we noticed that the bucket's end and start dates do not accurately reflect the interval range of the transactions stored in the array. By updating them, we obtain more precise and refined information about the date range.

- __EXPECTED RESULT:__ More precise data

```python
# Define the aggregation pipeline to compute the min and max transaction dates
pipeline = [
    # Unwind the transactions array to work with individual transaction documents
    {'$unwind': '$transactions'},
    
    # Group to find the min (oldest) and max (most recent) transaction dates
    {'$group': {
        '_id': '$_id',
        'min_transaction_date': {'$min': '$transactions.date'},
        'max_transaction_date': {'$max': '$transactions.date'}
    }},
    
    # Project to create new fields for bucket_start_date and bucket_end_date
    {'$project': {
        'bucket_start_date': '$min_transaction_date',
        'bucket_end_date': '$max_transaction_date'
    }},
]

result = db.Transactions.aggregate(pipeline)
```

```python
for document in result:
    new_bucket_end_date = document['bucket_end_date']
    new_bucket_start_date = document['bucket_start_date']
    
    result = db.Transactions.update_one(
        {'_id': document['_id']},
        {'$set': {'bucket_end_date': new_bucket_end_date, 'bucket_start_date': new_bucket_start_date}}
    )
```

### Question 7

This MongoDB aggregation pipeline filters and groups listings by property_type and city, calculating the average maximum_nights. It then sorts the results to identify the cities with the highest and lowest average maximum_nights per property type, presenting this data in a structured format for analysis and reporting.

```python
pipeline = [
    # Match documents with valid maximum_nights, property_type, and city
    {"$match": {
        "maximum_nights": {"$exists": True, "$ne": None},
        "property_type": {"$exists": True, "$ne": None},
        "address.market": {"$exists": True, "$ne": None}
    }},
    
    # Convert maximum_nights to integer for aggregation
    {"$addFields": {
        "maximum_nights_int": {"$toInt": "$maximum_nights"}
    }},
    
    # Group by property_type and city, calculate average maximum_nights
    {"$group": {
        "_id": {
            "property_type": "$property_type",
            "city": "$address.market"
        },
        "avg_max_nights": {"$avg": "$maximum_nights_int"}
    }},
    
    # Sort by property_type and descending average maximum_nights
    {"$sort": {
        "_id.property_type": 1,
        "avg_max_nights": -1
    }},
    
    # Group again by property_type to find max and min avg_max_nights per property_type
    {"$group": {
        "_id": "$_id.property_type",
        "cities": {
            "$push": {
                "city": "$_id.city",
                "avg_max_nights": "$avg_max_nights"
            }
        },
        "max_city": {"$first": {"city": "$_id.city", "avg_max_nights": "$avg_max_nights"}},
        "min_city": {"$last": {"city": "$_id.city", "avg_max_nights": "$avg_max_nights"}}
    }},
    
    # Project the final output
    {"$project": {
        "_id": 0,
        "property_type": "$_id",
        "max_city": 1,
        "min_city": 1
    }}
]

# Execute the aggregation pipeline
result = list(db.listingsAndReviews_new.aggregate(pipeline))
```

- __TRANSFORMATION APPLIED:__ The transformation applied involves restructuring the existing document schema to adopt the Attribute Pattern by introducing a new field called "METRICS". This field is an array of sub-documents, where each sub-document contains a name and a score field. With this pattern, querying across all metrics to compute averages becomes simpler and more efficient.

- __EXPECTED RESULT:__ With the implemented changes, future queries intended to calculate the average review scores across all metrics (including any newly added ones) can be executed without modifying the query itself.

Step 2: Choose a random city from the Top 10 Cities, randomly choose its details, generate property details based on their distribution in said location.

```python
def get_location_details(city_name):
    """
    Retrieves a series of location details including suburb, street, government area, and country information for a specified city.
    """
    # Get random suburb
    suburb_pipeline = [
        {'$match': {'address.market': city_name}},
        {'$group': {'_id': '$address.suburb'}},
        {'$sort': {'_id': 1}},
        {'$project': {'suburb': '$_id', '_id': 0}}
    ]
    suburbs = [item['suburb'] for item in db.listingsAndReviews_new.aggregate(suburb_pipeline) if item['suburb']]
    suburb = choice(suburbs) if suburbs else ''

    # Get random street
    street_pipeline = [
        {'$match': {'address.market': city_name, 'address.suburb': suburb}},
        {'$group': {'_id': '$address.street'}},
        {'$sort': {'_id': 1}},
        {'$project': {'street': '$_id', '_id': 0}}
    ]
    streets = [item['street'] for item in db.listingsAndReviews_new.aggregate(street_pipeline) if item['street']]
    street = choice(streets) if streets else ''

    # Get government area
    gov_area_pipeline = [
        {'$match': {'address.street': street}},
        {'$group': {'_id': '$address.government_area'}},
        {'$sort': {'_id': 1}},
        {'$project': {'government_area': '$_id', '_id': 0}}
    ]
    gov_areas = [item['government_area'] for item in db.listingsAndReviews_new.aggregate(gov_area_pipeline) if item['government_area']]
    government_area = gov_areas[0] if gov_areas else ''

    # Get country
    country_pipeline = [
        {'$match': {'address.street': street, 'address.government_area': government_area}},
        {'$project': {'country': '$address.country', 'country_code': '$address.country_code', '_id': 0}},
        {'$limit': 1}
    ]
    result = next(db.listingsAndReviews_new.aggregate(country_pipeline), None)
    country_info = (result['country'], result['country_code']) if result else None

    return {
        'suburb': suburb,
        'street': street,
        'government_area': government_area,
        'country': country_info[0] if country_info else None,
        'country_code': country_info[1] if country_info else None
    }

def generate_property_details(street):
    """
    Generates both the property type and bed count based on the most common occurrences found on a given street.
    """
    # Generate the most common property type
    property_type_pipeline = [
        {'$match': {'address.street': street}},
        {'$group': {'_id': '$property_type', 'count': {'$sum': 1}}},
        {'$sort': {'count': -1}},  # Sort by occurrence to get the most common
        {'$project': {'property_type': '$_id', '_id': 0}},
        {'$limit': 1}
    ]
    property_type_result = next(db.listingsAndReviews_new.aggregate(property_type_pipeline), None)
    property_type = property_type_result['property_type'] if property_type_result else None

    # Generate the most common number of beds
    beds_pipeline = [
        {'$match': {'address.street': street, 'property_type': property_type}},  # Filter by street and the most common property type
        {'$group': {'_id': '$beds', 'count': {'$sum': 1}}},
        {'$sort': {'count': -1}},  # Sort by occurrence to get the most common
        {'$project': {'beds': '$_id', '_id': 0}},
        {'$limit': 1}
    ]
    beds_result = next(db.listingsAndReviews_new.aggregate(beds_pipeline), None)
    beds = beds_result['beds'] if beds_result else None

    return {
        'property_type': property_type,
        'number_of_beds': beds
    }
```

### Question 15

We added a new review metric called "x_factor" to a specific listing document in the listingsAndReviews_new collection. The "x_factor" metric has a score of 10. After adding this new metric, we recalculated the average score across all metrics for the same listing using the previously developed query.

```python
# Calculating the average score for the property with ID "100065461" across all review metrics.

pipeline = [
    {"$match": {"_id": "100065461"}},  
    {"$unwind": "$METRICS"},
    {"$group": {"_id": "$_id", "averageScore": {"$avg": "$METRICS.score"}}},
    {"$project": {"_id": 1, "averageScore": 1}}
]

result = db.listingsAndReviews_new.aggregate(pipeline)

pprint(list(result))
```

```python
# Adding a new review metric called "x_factor" with a score of 10 to the document.

db.listingsAndReviews_new.update_one(
    {'_id': '100065461'},
    {'$set': {'METRICS': [
        {"name": "check-in", "score": None},
        {"name": "cleanliness", "score": None},
        {"name": "communication", "score": None},
        {"name": "location", "score": None},
        {"name": "rating", "score": None},
        {"name": "value", "score": None},
        {"name": "x_factor", "score": 10}  
    ]}}
)
```
